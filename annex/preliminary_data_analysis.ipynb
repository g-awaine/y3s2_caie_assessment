## Null Check
Investigates the missing values identified in the Overview section.

```python
# Calculates the percentage of null entries in each column i.e. number of null entries / number of entries in the column
def calculate_null_proportion_per_column(df):
    num_null_entries_per_column = df.isnull().sum()
    total_rows = df.shape[0]
    null_proportion = num_null_entries_per_column/total_rows *100
    print("% of null entries per column:")
    return null_proportion.round(2)

# Calculates the number and percentage of rows with at least one null entry
def count_rows_with_nulls(df):
    # Count rows with at least one null
    num_rows_with_nulls = df.isnull().any(axis=1).sum()
    total_rows = df.shape[0]
    percent_rows_with_null = (num_rows_with_nulls / total_rows) * 100  # Calculate percentage

    # Display results
    print(f"No. rows with at least one null value: {num_rows_with_nulls}")
    print(f"% of rows with at least one null value: {percent_rows_with_null:.2f}%")
```

The following datasets have missing values and will be the only ones investigated in this section

The other datasets do not have null values in them and will not need to be investigated for null values.

## Duplicate Check

```python
# Calculates the number and proportion of rows which are duplicate entries
def duplicate_entries_search(df):
    # Identify total number of rows in the dataset
    total_rows = df.shape[0]

    # Count duplicate rows
    num_duplicate_rows = df.duplicated().sum()
    percent_duplicate_rows = (num_duplicate_rows / total_rows) * 100

    # Display results
    print(f"Total duplicate rows: {num_duplicate_rows}")
    print(f"% of duplicate rows: {percent_duplicate_rows:.2f}%")


# Calculates the number and proportion of rows which have duplicate values in a specified data field
def duplicate_values_search(df, field):
    # Identify total number of rows in the dataset
    total_rows = df.shape[0]

    # Count duplicate values in the data field
    num_duplicate_values = df[field].duplicated().sum()
    proportion_duplicate_values = (num_duplicate_values / total_rows) * 100

    # Display results
    print(f"\n{field}:")
    print(f"Total duplicate entries: {num_duplicate_values}")
    print(f"% of duplicate entries: {proportion_duplicate_values:.2f}%")
```

```python
# Check if the given text contains any diacritic characters.
def identify_diacritics(text):
    for char in text:
        # Normalize the character to NFD and check for combining marks
        if any(unicodedata.category(mark) == 'Mn' for mark in unicodedata.normalize('NFD', char)):
            return True
    return False

# Identify characters in the text that are not letters or diacritics.
def identify_non_letter_and_non_diacritic(text):
    # Initialise the non diacritic characters list
    non_letter_diacritic_chars = []
    
    for char in text:
        # Get the Unicode category of the character
        char_category = unicodedata.category(char)
        # 'L' = Letter categories, 'M' = Mark categories (diacritics)
        if not char_category.startswith(('L', 'M')) and char not in [' ', '-', '\'', '(', ')']:
            non_letter_diacritic_chars.append(char)
    return bool(non_letter_diacritic_chars)
```

```python
# Identify any diacritics in the customers' cities
customers_df[customers_df['customer_city'].apply(identify_diacritics)]
```

```python
# Identify text that may have encoding errors
customers_df[customers_df['customer_city'].apply(identify_non_letter_and_non_diacritic)]
```

```python
# Merge with items dataset to identify if the payment values adds up to the products price and freight price
missing_payment_sequential_df = missing_payment_sequential_df.merge(items_df, how='left', on='order_id')

# Combine the price and freight value the different products for each unique order
missing_payment_sequential_df = missing_payment_sequential_df.groupby('order_id').agg({
    'payment_value': lambda x: x.mode()[0] if not x.mode().empty else None,
    'price': 'sum',
    'freight_value': 'sum'
})

# Sum up the price and freight value and compare it with the payment_value
missing_payment_sequential_df['total_order_cost'] = missing_payment_sequential_df['price'] + missing_payment_sequential_df['freight_value']
missing_payment_sequential_df
```
